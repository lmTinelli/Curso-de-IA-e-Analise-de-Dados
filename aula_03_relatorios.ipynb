## Maritaca

## A idade NÃO explica notas mais altas — é um efeito ilusório

Essa é uma excelente pergunta, porque toca num ponto clássico da análise de dados: **a diferença percebida pelo olhar humano nem sempre existe nos números**. Vamos aos fatos desta base:

---

### 1. Correlação quase zero

A correlação entre idade e nota final é de **−0,01** (p = 0,75) — estatisticamente **indistinguível de zero**. Ou seja, não existe relação linear entre ser mais velho e tirar nota maior.

---

### 2. Média por faixa etária — sem padrão

| Faixa etária | Nota média |
|---|---|
| 18–20 anos | **69,7** |
| 21–23 anos | **69,2** |
| 24+ anos | **69,6** |

A diferença entre os grupos é de **menos de 1 ponto** — irrelevante.

---

### 3. Mesmo controlando por horas de estudo, nada muda

Dividindo os alunos em tercis de estudo (baixo, médio, alto) e medindo a correlação idade × nota **dentro de cada grupo**:

| Nível de estudo | Correlação idade × nota | p-valor |
|---|---|---|
| Baixo | −0,017 | 0,756 |
| Médio | +0,022 | 0,685 |
| Alto | +0,019 | 0,737 |

**Em todos os casos, a correlação é zero.** Isso elimina a hipótese de que os mais velhos tiram notas melhores porque estudam mais.

---

### 4. Por que a impressão existe então?

O que provavelmente aconteceu é um **viés de confirmação** ou **efeito de pequenos números**:

- Na faixa dos 23–24 anos, há algumas notas muito altas (100, 97, 95...) que chamam a atenção — mas também há notas baixas. A **média** permanece igual.
- Com apenas 134 alunos de 24 anos, um punhado de outliers com nota 100 cria a **impressão** de que "os mais velhos vão melhor", mesmo sem padrão real.
- O gráfico de dispersão confirma: as notas estão espalhadas aleatoriamente pela idade, sem tendência ascendente ou descendente.

---

### Conclusão

> **Idade não é um preditor de desempenho nesta base.** A diferença observada é fruto do acaso e de flutuação amostral — não de um mecanismo real. Os fatores que de fato importam (e que já analisei antes) são: **horas de estudo** (correlação 0,81), **saúde mental** (0,32) e **tempo em redes sociais** (−0,17). É neles que a escola deve concentrar atenção.
